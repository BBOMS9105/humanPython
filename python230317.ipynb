## 데이터 핸들링
- 행과 열을 다루는 방법

```python
lemonade.head()
```

- 새 컬럼 추가

```python
lemonade['Sold'] = 0
lemonade.head()
```

```python
lemonade['Sold'] = lemonade['Lemon'] + lemonade['Orange']
lemonade.head(3)
```

- Revenue 컬럼 : 매출 계산

```python
lemonade['Revenue'] = lemonade['Sold'] * lemonade['Price']
lemonade.head()
```

```python
lemonade['Temp C'] = (lemonade['Temperature'] - 32) * 0.5556
lemonade.head()
```

```python
lemonade['Temp C'] = lemonade['Temp C'].round(1)
lemonade.head()
```

- 특정 컬럼 삭제

```python
lemonade_col_drop = lemonade.drop('Sold', axis = 1)
lemonade_col_drop.head()
```

```python
# 행 삭제
lemonade_row_drop = lemonade.drop(0, axis = 0)
lemonade_row_drop.head()
```

- 행 삭제 후 인덱스 번호 초기화
  + 인덱스가 0부터 다시 시작하도록 변환

## 데이터 추출
- 조건식을 사용하여 데이터 추출
- 조건식에서 참(True)인 값만 추출

- 행과 열을 동시처리 --> loc만 처리

```python
#                        행 추출 조건식              열 추출             인덱스 초기화
result = lemonade.loc[lemonade['Revenue'] > 100, ['Date', 'Revenue']].reset_index(drop = True)
result
```

```python
# 날짜와 총 판매량, 기온, 단가, 수익 추출
# 장소는 beach
lemonade.loc[lemonade['Location'] == 'Beach', ['Date', 'Temp C', 'Sold', 'Price', 'Revenue']].reset_index(drop = True)
```

```python
# 날짜와 총 판매량, 기온, 단가, 수익 추출
# 장소는 beach
cols = ['Date', 'Temp C', 'Sold', 'Price', 'Revenue']
lemonade.loc[lemonade['Location'] == 'Beach', cols].reset_index(drop = True)
```

### 집계함수(Aggreegation)
- Groupby 및 집계 --> 피봇테이블

```python
import numpy as np
lemonade.groupby('Location')['Revenue'].agg([max, min, np.mean])
```

- specis 컬럼이 setosa인 것만 전체 조회

```python
result = iris[iris['species'] == 'setosa'].reset_index(drop = True)
result.head()
```

- sepal_width < 3보다 작은 데이터 조회

```python
result2 = result[result['sepal_width'] < 3].reset_index(drop = True)
result2
```

- AND &
- OR  |

```python
# iris[(조건식1) & (조건식2) & (조건식3)] <-- 이런식으로 만들고 조건식을 작성해야 헷갈리지 않는다
```

```python
# result3 = iris[(조건식1) & (조건식2)]
result3 = iris[(iris['species']=='setosa') & (iris['sepal_width'] < 3)].reset_index(drop=True)
result3
```
